```python
import numpy as np
import pandas as pd
```

```python
#Reading in the csv file data as a pandas DataFrame
netflix = pd.read_csv('netflix_titles.csv')
```

```python
#Filtering out netflix movies as we are focusing on TV Shows
length = len(netflix['show_id'])
for i in range(0,length):
    if netflix.loc[i, 'type'] != 'TV Show':
        netflix = netflix.drop(index=i)
netflix = netflix.reset_index(drop=True)
netflix
```

```python
#We want to keep
#show_id, title, listed_in, description, release_year, cast  
shows = netflix[['show_id','title', 'cast','listed_in','description']]
shows
```

```python
#New dataframe without shows that don't have a cast
shows_final = shows.dropna().reset_index(drop=True)
shows_final
```

```python
shows_final.duplicated().sum()
```

```python
#We convert the cast, listed_in and description fields into list format
shows_final['description'] = shows_final['description'].apply(lambda x:x.split()) #splits on space automatically
shows_final['cast'] = shows_final['cast'].apply(lambda x:x.split(','))
shows_final['listed_in'] = shows_final['listed_in'].apply(lambda x:x.split(','))
shows_final
```

```python
#To minimize confusion (eg. if actors have same first names) we eliminate spaces in the elements
#so we compare keyword ChrisHemsworth instead of keyword Chris and 2nd keyword Hemsworth in model
#Note: model takes a string full of keywords, so that is why we do this

shows_final['cast'] = shows_final['cast'].apply(lambda x:[i.replace(" ","") for i in x])
shows_final['listed_in'] = shows_final['listed_in'].apply(lambda x:[i.replace(" ","") for i in x])
shows_final['cast'].apply(lambda x:[i.replace(" ","") for i in x])
shows_final
```

```python
#Concatinating the three seperate keyword columns into one column (one big list)
shows_final['keywords'] = shows_final['cast'] + shows_final['listed_in'] + shows_final['description']
shows_final = shows_final[['show_id', 'title', 'keywords']]
shows_final
```

```python
#For purposes of our model we convert keywords list to string seperated by space
# we also make it all lowercase so keyword matching for description portion is not case sensitive
shows_final['keywords'] = shows_final['keywords'].apply(lambda x:" ".join(x))
shows_final
```

```python
#Analyzing the keywords list we see some issues
#Can we remove grammatical redundencies (eg. activity, activities)

import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    lst = []
    for i in text.split():
        lst.append(ps.stem(i)) #Changes grammatical redundency to base word 
        #eg. activities -> activity, funner -> fun
    return " ".join(lst) #adds stem to the final string
        
```

```python
shows_final['keywords'] = shows_final['keywords'].apply(stem)
```

```python
#Now the strategy is to take this major text in keywords and make a unique list (top n words)
#For each show we make a frequency list on how many times each word in unique list
#shows up in that show's keywords
#Now we make a n x n df where columns are the words in unique list and rows are shows
#so we have a vector vertically for each word and how much it appears in each show
#we have a vector horizontally for each show and how much each word shows up in it

##Note we have to avoid words like in, a, are, is, etc. 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#We will use a giant maximum words to take all the keywords from different shows
#We create the cv object from the CountVectorizer class
cv = CountVectorizer(max_features=1000000,stop_words='english')
#Here we use the countvectorizer on the keywords to create a df as described in above cell
#Note that shows_final['keywords'] is a list of the keyword strings.
#fit_transform method takes all words and counts how many times each keyword appears in each eleement of the array
vectors = cv.fit_transform(shows_final['keywords']).toarray()
#For each keyword we have a frequency list on how many times that word appears in each show
#The vertical vectors would be how many times each show holds the keyword
vectors
```

```python
keywords = cv.get_feature_names() #all words
len(keywords)
keywords
```

```python
#Now that we have a refined list we just have to do the comparison
#With high-dimensional space, tip-to-tip distance between vectors is not the best measure
#We instead use the cosine distance (angle between two vectors)
#For each movie we check it's cosine distance with each other movie
#The higher the cosine distance the lower the similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#cosine_similarity is like inverse of cosine_distance,
#directly gives us similarity

show_similarities = cosine_similarity(vectors)
#This show_similarities matrix holds the similarity of each show with all the other 
#shows based on the vectors array. For each show there is a list with it's similarity ratings to each show.
show_similarities
```

```python
#Given the show_similarities matrix, we just want to get the most similar show
#to the given show. (In the matrix vector for given show, return top 10 highest similarity ratings)
#besides the show's perfect similarity rating to itself.
def show_reccommendation(movie):
    #This gets the index of the given movie in the shows df (0th column is index column)
    show_index = shows_final[shows_final['title'] == movie].index[0]
    
    #This gets the similarities of given movie using it's index
    #We get this list from the similarities matrix with the index'th
    #row being the similarity array of the given movie
    similarities = show_similarities[show_index]
    
    #we get a sorted list of similarities of given movie, taking 2nd-11th most similar (1st would be itself)
    #btw, list(enumerate()) makes a list of tuples with original index, value
    #here we extraxt the tuples of top 10 most similar
    top_sims = sorted(list(enumerate(similarities)), reverse=True, key=lambda x:x[1])[1:11]
    
    #for each similarity tuple, i[0] is the original index of the movie.
    #We use this to get from shows_final, the title of the movie with that index
    for i in top_sims:
        print(shows_final.iloc[i[0]].title)

#How can we get the index of a movie based on it's
```
